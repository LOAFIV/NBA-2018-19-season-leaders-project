```python
#Webscrapping NBA.com stats for 2018-19 season with API  using Pandas and Requests libraries only. 
```

```python
#Impport necessary libraries
import pandas as pd
import requests
```

```python
#Locate API url on NBA.com website
test_url = 'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season=2018-19&SeasonType=Regular%20Season&StatCategory=PTS'
```

```python
#download json file containing required data using requests library
r = requests.get(url=test_url).json()
```

```python
# pull required data and headers using requests library
table_headers = r['resultSet']['headers']
```

```python
# create data frame using Pandas library to store data, assign headers to the columns
pd.DataFrame(r['resultSet']['rowSet'], columns = table_headers)
```

```python
#create a for loop to add year and season_type columns to data set. This could be useful if multiple years or seasons are to be added to the dataset.
temp_df1 = pd.DataFrame(r['resultSet']['rowSet'], columns = table_headers)
temp_df2 = pd.DataFrame({'Year':['2018-19' for i in range(len(temp_df1))],
                         'Season_type':['Regular%20Season' for i in range(len(temp_df1))]})

 #combine both temp dataframes. Columns for Year and Season Type added
temp_df3 = pd.concat([temp_df2,temp_df1], axis=1) 
temp_df3
```

```python
#Export raw/dirty data to CSV file.
temp_df3.to_csv("C:\\Users\\LENOVO\\Downloads\\nba_leaders_stats_2018-19.csv", index = False)
```
